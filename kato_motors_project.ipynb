**Dropping Unwanted Columns:**

Further after Analysing, I find that car_ID is nothing but a range of numbers from 1-205 this doesn't impact the price of the car and  columns named Untitled and Unnamed: 0 which have only null values. Hence, I begin by removing unnecessary columns from my dataset.
```python
df.drop(['Unnamed: 0','car_ID','Untitled'], axis = 1, inplace = True)
df.head()
```

**Handling Missing Values:**

Next, I address missing values in dataset. Missing values can impact the accuracy of my analysis, so I take the following steps:
For the 'symboling' column, I fill missing values with the mean of the column.
For the 'price' column, we fill missing values with the median.
For the 'fueltype' column, we fill missing values with the most frequent value (mode)



---



```python
df['symboling'].fillna(df['symboling'].mean(), inplace=True)
df['price'].fillna(df['price'].median(), inplace=True)

most_frequent_fueltype = df['fueltype'].mode()[0]
df['fueltype'].fillna(most_frequent_fueltype, inplace=True)


print(df.isnull().sum())
```

**Identifying Outliers:**

To detect outliers, I calculate quartiles (Q1 and Q3) and the Interquartile Range (IQR) for the 'price' column. Outliers are then identified as values falling below the lower bound (Q1 - 1.5 * IQR) or above the upper bound (Q3 + 1.5 * IQR).

---
**Applying a Log Transformation to Handle Outliers:**

To mitigate the impact of outliers on analysis, I applied a log transformation to the 'price' column. Logarithmic transformations are a common technique to make data more normally distributed and reduce the influence of extreme values.


---






```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
print("\nOutliers in price:")
print(outliers)
```

The 'CarName' column in dataset contains valuable information about car models and their manufacturers. However, to ensure data uniformity and eliminate potential discrepancies, I will extract and store the car manufacturers' names in a new column called 'car_company.' This transformation will enable us to analyze and group data more effectively based on manufacturers.


---



```python
car_companies = pd.Series([car.split(" ")[0] for car in df['CarName']], index = df.index)
print(car_companies.head(25))
```

**Drop 'CarName' Column:**

To maintain a cleaner dataset and avoid redundancy, I removed the 'CarName' column from the DataFrame. I have already extracted the relevant information into 'car_companies.'


---



```python
df.drop(columns= ['CarName'], axis = 1, inplace = True)
```

**Create a New 'car_company' Column:**

This step introduce a new column named 'car_company' to the DataFrame, storing the standardized car manufacturer names extracted earlier.


---



```python
df['car_company'] = car_companies
print(df['car_company'].value_counts())
```

**Replace Misspelt 'car_company' Names:**

Notice that the following manufacturer names are misspelt or not regularised (follows the same case or uppercase or lower case or title case):

vw and vokswagen should be volkswagen,
porcshce should be porsche,
toyouta should be toyota,
Nissan should be nissan,
maxda should be mazda

To ensure consistency and accuracy, I identify and replace misspelled car manufacturer names with their correct counterparts.


---



```python
df.loc[(df['car_company'] == "vw") | (df['car_company'] == "vokswagen"), 'car_company'] = 'volkswagen'
df.loc[df['car_company'] == "porcshce", 'car_company'] = 'porsche'
df.loc[df['car_company'] == "toyouta", 'car_company'] = 'toyota'
df.loc[df['car_company'] == "Nissan", 'car_company'] = 'nissan'
df.loc[df['car_company'] == "maxda", 'car_company'] = 'mazda'
print(df['car_company'].value_counts())
```

```python
company_counts = df['car_company'].value_counts()
from matplotlib import colormaps
list(colormaps)

# Plot bar chart
plt.figure(figsize=(10, 6))
company_counts.plot(kind='bar', color= 'Purple')

# Add labels and title
plt.xlabel('Car Company')
plt.ylabel('Count')
plt.title('Count of Cars by Company')

# Show plot
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

This phase of project is crucial, as it involves identifying the key factors that have a substantial impact on car prices, setting the stage for the development of a robust pricing model. Utilising powerful data analysis tools such as Pandas and NumPy, I will uncover correlations and relationships of all the numerical columns with the price of cars.


---



```python
numerical_columns = df.select_dtypes(include=[np.number])
correlation_matrix = numerical_columns.corr()
price_correlations = correlation_matrix['price']
print(f"\nCorrelation coefficients with 'price':\n{price_correlations}")
```

In the final step of this phase, I set a threshold of 0.5, beyond which correlation coefficients are considered significant. I identify and list the attributes that meet this criterion. These factors are the key influencers of car prices in dataset.


---



```python
significant_factors = price_correlations[np.abs(price_correlations) > 0.5]
print(f"\nFactors with significant impact on car prices:\n{significant_factors}")
```

```python
# Plot bar chart
plt.figure(figsize=(8,5))
significant_factors.plot(kind='bar', color='skyblue')

# Add labels and title
plt.xlabel('Factors')
plt.ylabel('Correlation Coefficient')
plt.title('Factors with Significant Impact on Car Prices')

# Show plot
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

**Summary**

By executing these steps, I've successfully uncovered the Numerical factors that wield substantial influence over car prices in the American market. These insights will guide subsequent analysis and model development, steering us toward a predictive pricing model that aligns with our ultimate goalâ€”Kato Motors' competitive edge in the automotive landscape.

**Identifying Categorical Factors Impacting Car Prices:**

In final process to identify the factors impacting the prices, I have successfully found the numerical values influencing the price. Now, I turn my attention to the categorical variables present in dataset. These variables encapsulate essential attributes, such as the type of fuel, aspiration, car body style, and drivetrain, which can significantly impact the pricing of cars.


---



```python
categorical_columns = df.select_dtypes(include=['object','category'])
for column in categorical_columns:
    print(f"Impact of '{column}' on 'price':")
    unique_categories = df[column].unique()
    for category in unique_categories:
        category_data = df[df[column] == category]
        price_summary = category_data['price'].describe()
        print(f"\nCategory: {category}")
        print(price_summary)
    print('\n')
```

**Summary**

Through this meticulous examination of categorical variables and their correlation with 'price,' I've unveiled valuable insights. These insights provide with a nuanced understanding of how attributes like fuel type, aspiration, car body style, and drivetrain influence car pricing in the American market.
