**Dropping Unwanted Columns:**

Further after Analysing, I find that car_ID is nothing but a range of numbers from 1-205 this doesn't impact the price of the car and  columns named Untitled and Unnamed: 0 which have only null values.
```python
df.drop(['Unnamed: 0','car_ID','Untitled'], axis = 1, inplace = True)
df.head()
```

**Handling Missing Values:**

Next, I address missing values in dataset. Missing values can impact the accuracy of my analysis, so I take the following steps:
For the 'symboling' column, I fill missing values with the mean of the column.
For the 'price' column, we fill missing values with the median.
For the 'fueltype' column, we fill missing values with the most frequent value (mode)



---



```python
df['symboling'].fillna(df['symboling'].mean(), inplace=True)
df['price'].fillna(df['price'].median(), inplace=True)

most_frequent_fueltype = df['fueltype'].mode()[0]
df['fueltype'].fillna(most_frequent_fueltype, inplace=True)


print(df.isnull().sum())
```

**Identifying Outliers:**

To detect outliers, I calculate quartiles (Q1 and Q3) and the Interquartile Range (IQR) for the 'price' column. Outliers are then identified as values falling below the lower bound (Q1 - 1.5 * IQR) or above the upper bound (Q3 + 1.5 * IQR).

---
**Applying a Log Transformation to Handle Outliers:**

To mitigate the impact of outliers on analysis, I applied a log transformation to the 'price' column. Logarithmic transformations are a common technique to make data more normally distributed and reduce the influence of extreme values.


---






```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['price'] < lower_bound) | (df['price'] > upper_bound)]
print("\nOutliers in price:")
print(outliers)
```

The 'CarName' column in dataset contains valuable information about car models and their manufacturers. However, to ensure data uniformity and eliminate potential discrepancies, I will extract and store the car manufacturers' names in a new column called 'car_company.' This transformation will enable us to analyze and group data more effectively based on manufacturers.


---



```python
car_companies = pd.Series([car.split(" ")[0] for car in df['CarName']], index = df.index)
print(car_companies.head(25))
```

**Drop 'CarName' Column:**

To maintain a cleaner dataset and avoid redundancy, I removed the 'CarName' column from the DataFrame. I have already extracted the relevant information into 'car_companies.'


---



```python
df.drop(columns= ['CarName'], axis = 1, inplace = True)
```

**Create a New 'car_company' Column:**

This step introduce a new column named 'car_company' to the DataFrame, storing the standardized car manufacturer names extracted earlier.


---



```python
df['car_company'] = car_companies
print(df['car_company'].value_counts())
```

**Replace Misspelt 'car_company' Names:**

Notice that the following manufacturer names are misspelt or not regularised (follows the same case or uppercase or lower case or title case):

vw and vokswagen should be volkswagen,
porcshce should be porsche,
toyouta should be toyota,
Nissan should be nissan,
maxda should be mazda

To ensure consistency and accuracy, I identify and replace misspelled car manufacturer names with their correct counterparts.


---



```python
df.loc[(df['car_company'] == "vw") | (df['car_company'] == "vokswagen"), 'car_company'] = 'volkswagen'
df.loc[df['car_company'] == "porcshce", 'car_company'] = 'porsche'
df.loc[df['car_company'] == "toyouta", 'car_company'] = 'toyota'
df.loc[df['car_company'] == "Nissan", 'car_company'] = 'nissan'
df.loc[df['car_company'] == "maxda", 'car_company'] = 'mazda'
print(df['car_company'].value_counts())
```
